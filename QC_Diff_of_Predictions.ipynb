```
import pandas as pd

df_Model = pd.read_csv('./csv/submission.csv')
df_Transfer = pd.read_csv('./csv/submission_from_Transfer.csv')
```

```
diff = 0
diff_Model = []
diff_Transfer = []
for i in range(0, 19278):
    if df_Model.iloc[i][1] != df_Transfer.iloc[i][1]:
        diff_Model.append(df_Model.iloc[i][1])
        diff_Transfer.append(df_Transfer.iloc[i][1])
        diff += 1

diff
```

```
import numpy as np

# 混淆矩阵
table=pd.crosstab(np.array(diff_Model), np.array(diff_Transfer), colnames=['Model'], rownames=['Transfer'])
table
```
